```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('for_loop_result.tsv') as f:
    data = f.readlines()
```

```python
string = ''.join(data)
data = string.split('\n')
data.pop()
data = np.array(data).reshape(8, 2)
data
```

```python
fig, ax = plt.subplots()
ax.plot(data[:,0], data[:,1].astype(float), 'o-')
ax.set(xlabel='Python version', ylabel='Time (s)',
         title='For loop performance (1e+8 iterations)')
ax.grid()
fig.savefig("for_loop_performance.png")
```
